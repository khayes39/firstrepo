```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import make_pipeline

drive.mount('/content/drive')
file_path = '/content/drive/My Drive/ML/Data_sets/Housing.csv'
housing = pd.DataFrame(pd.read_csv(file_path))
print(housing.shape)
housing.head()
```

```python
def binary_map(x):
    return x.map({'yes': 1, 'no': 0})

binary_varlist =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
# Applying the function to the housing list
housing[binary_varlist] = housing[binary_varlist].apply(binary_map)
housing.head()
```

```python
# x_vars = ["area", "bedrooms", "bathrooms", "stories", "mainroad", "guestroom", "basement", "hotwaterheating", "airconditioning", "parking", "prefarea"]

#splitting into x and y data
x_vars = np.arange(1, 12, 1)
X_vals = housing.iloc[:, x_vars].values
Y_vals = housing.iloc[:, 1].values



#standardization of the inputs!
scaler = StandardScaler()
x_std = scaler.fit_transform(X_vals)
# print(x_std)

#splitting into test and train
np.random.seed(0)
x_train, x_test, y_train, y_test = train_test_split(x_std, Y_vals, train_size = 0.8, test_size = 0.2, random_state = 100)
# print(x_train)
```

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
num_features = x_std.shape[1]




for kernel in kernels:  #trying with each kernel

    if kernel == 'linear':
        svr = SVR(kernel='linear', C=1e3)
    elif kernel == 'poly':
        svr = SVR(kernel='poly', C=1e3, degree=2)
    elif kernel == 'rbf':
        svr = SVR(kernel='rbf', C=1e3, gamma=0.1)
    elif kernel == 'sigmoid':
        svr = SVR(kernel='sigmoid', C=1e3, gamma=0.1)
    mse_array = []

    print("iterating through n features for kernel: ", kernel)

    for n in range(1, num_features+1):
        pca = PCA(n_components=n)
        x_train_pca = pca.fit_transform(x_train)
        x_test_pca = pca.transform(x_test)

        y = svr.fit(x_train_pca, y_train).predict(x_test_pca)

        mse = mean_squared_error(y_test, y)
        print("for kernel: ", kernel, " and ", n, "features, the MSE is: ", mse)

        mse_array.append(mse)

        #get an array of MSE values here and then plot after this for loop. one plot for each kernel
    plt.figure()
    plt.plot(range(1,num_features+1), mse_array)
    plt.title("MSE vs N components over kernel: "+kernel)
    plt.xlabel("n PCA components")
    plt.ylabel("Mean Squared Error")
    plt.show()



```
